````python
@tool
def text_to_aql_to_text(query: str):
    """Translate natural language queries into AQL and execute them."""
    # Step 1: Generate AQL from natural language query using DeepSeek via OpenRouter
    aql_prompt = f"""
    I have an ArangoDB database with the following collections:
    - Document Collections: patients, allergies, conditions, encounters, etc.
    - Edge Collections: patients_to_allergies, patients_to_conditions, encounters_to_allergies, etc.

    Translate the following natural language query into an ArangoDB AQL query:
    Query: {query}

    Return only the AQL query. Do not include any explanations or additional text.
    """
    
    aql_response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model on OpenRouter
        messages=[
            {"role": "user", "content": aql_prompt}
        ]
    )
    
    aql_query = aql_response.choices[0].message.content.strip()
    print(f"Generated AQL Query: {aql_query}")
    
    # Step 2: Execute the AQL query on ArangoDB
    try:
        cursor = db.aql.execute(aql_query)
        query_results = list(cursor)
    except Exception as e:
        return f"Error executing AQL query: {e}"
    
    # Step 3: Translate query results into natural language using DeepSeek via OpenRouter
    results_prompt = f"""
    Translate the following ArangoDB query results into natural language:
    Query: {query}
    Results: {query_results}

    Provide a concise and user-friendly response.
    """
    
    results_response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model on OpenRouter
        messages=[
            {"role": "user", "content": results_prompt}
        ]
    )
    
    natural_language_result = results_response.choices[0].message.content.strip()
    return natural_language_result
````

````python
import nx_cugraph as nxcg  # Import NetworkX-cuGraph

@tool
def text_to_nx_algorithm_to_text(query: str):
    """This tool is available to invoke a NetworkX Algorithm on
    the ArangoDB Graph. You are responsible for accepting the
    Natural Language Query, establishing which algorithm needs to
    be executed, executing the algorithm, and translating the results back
    to Natural Language, with respect to the original query.

    If the query (e.g traversals, shortest path, etc.) can be solved using the Arango Query Language, then do not use
    this tool.
    """
    # Step 1: Generate NetworkX/cuGraph code from the natural language query
    print("1) Generating NetworkX/cuGraph code")

    code_generation_prompt = f"""
    I have a NetworkX Graph called `G_adb`. It has the following schema: {arango_graph.schema}

    I have the following graph analysis query: {query}.

    Generate the Python Code required to answer the query using the `G_adb` object.
    Use NetworkX or cuGraph algorithms as appropriate. Prefer cuGraph for large-scale or performance-critical tasks.

    Be very precise on the algorithm you select to answer this query. Think step by step.

    Only assume that networkx and nx_cugraph are installed, and other base python dependencies.

    Always set the last variable as `FINAL_RESULT`, which represents the answer to the original query.

    Only provide python code that I can directly execute via `exec()`. Do not include any instructions.

    Make sure that `FINAL_RESULT` stores a short & concise answer. Avoid setting this variable to a long sequence.

    Your code:
    """

    code_response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model on OpenRouter
        messages=[
            {"role": "user", "content": code_generation_prompt}
        ]
    )

    generated_code = code_response.choices[0].message.content.strip()
    generated_code_cleaned = re.sub(r"^```python\n|```$", "", generated_code, flags=re.MULTILINE).strip()

    print('-' * 10)
    print("Generated Code:")
    print(generated_code_cleaned)
    print('-' * 10)

    # Step 2: Execute the generated code
    print("\n2) Executing NetworkX/cuGraph code")
    global_vars = {"G_adb": G_adb, "nx": nx, "nxcg": nxcg}  # Add 'nxcg' for NetworkX-cuGraph
    local_vars = {}

    try:
        exec(generated_code_cleaned, global_vars, local_vars)
        FINAL_RESULT = local_vars.get("FINAL_RESULT", "No result found.")
    except Exception as e:
        print(f"EXEC ERROR: {e}")
        return f"EXEC ERROR: {e}"

    print('-' * 10)
    print(f"FINAL_RESULT: {FINAL_RESULT}")
    print('-' * 10)

    # Step 3: Translate the results into natural language
    print("3) Formulating final answer")

    results_translation_prompt = f"""
    I have a NetworkX Graph called `G_adb`. It has the following schema: {arango_graph.schema}

    I have the following graph analysis query: {query}.

    I have executed the following python code to help me answer my query:

    ---
    {generated_code_cleaned}
    ---

    The `FINAL_RESULT` variable is set to the following: {FINAL_RESULT}.

    Based on my original Query and FINAL_RESULT, generate a short and concise response to
    answer my query. Make the response actionable and relevant to healthcare professionals.

    Your response:
    """

    results_response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model on OpenRouter
        messages=[
            {"role": "user", "content": results_translation_prompt}
        ]
    )

    natural_language_result = results_response.choices[0].message.content.strip()
    return natural_language_result    
@tool
def hybrid_query_execution(query: str):
    """Execute hybrid queries combining AQL and NetworkX/cuGraph."""
    # Step 1: Use AQL for traversal
    traversal_result = text_to_aql_to_text(query)

    # Step 2: Use GPU-accelerated analytics for deeper analysis
    analysis_result = text_to_nx_algorithm_to_text(query)

    # Step 3: Combine results
    combined_result = {
        "traversal_result": traversal_result,
        "analysis_result": analysis_result
    }

    # Step 4: Translate combined results into natural language
    combined_prompt = f"""
    Translate the following combined query results into natural language:
    Query: {query}
    Traversal Results: {traversal_result}
    Analysis Results: {analysis_result}

    Provide a concise and actionable response.
    """

    combined_response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model on OpenRouter
        messages=[
            {"role": "user", "content": combined_prompt}
        ]
    )

    natural_language_result = combined_response.choices[0].message.content.strip()
    return natural_language_result
def classify_query(query: str) -> str:
    """
    Classify a query into SIMPLE, COMPLEX, or HYBRID using the DeepSeek API via OpenRouter.
    """
    # Define the prompt for query classification
    classification_prompt = f"""
    Classify the following query into one of these categories:
    - SIMPLE: Can be answered using AQL (e.g., traversals, filtering).
    - COMPLEX: Requires GPU-accelerated analytics (e.g., PageRank, community detection).
    - HYBRID: Combines AQL and GPU-accelerated analytics.

    Query: {query}

    Return only the category (SIMPLE, COMPLEX, or HYBRID).
    """

    # Call the DeepSeek API via OpenRouter
    response = client.chat.completions.create(
        model="deepseek/deepseek-r1:free",  # Use the DeepSeek model
        messages=[
            {"role": "user", "content": classification_prompt}
        ]
    )

    # Extract the classification from the response
    classification = response.choices[0].message.content.strip().upper()
    return classification
    
def query_agent(query: str):
    """Agent to classify and execute queries."""
    query_type = classify_query(query)
    if query_type == "SIMPLE":
        result = text_to_aql_to_text(query)
    elif query_type == "COMPLEX":
        result = text_to_nx_algorithm_to_text(query)
    elif query_type == "HYBRID":
        result = hybrid_query_execution(query)
    else:
        result = hybrid_query_execution(query)
    return result
````
